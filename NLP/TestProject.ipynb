```python
import spacy

# Load the saved model
nlp = spacy.load("movie_sentiment_model")
```

Analyze a anew review

```python
def predict_sentiment(text):
    doc = nlp(text)
    sentiment = max(doc.cats, key=doc.cats.get)
    score = doc.cats[sentiment]
    return sentiment, score

# Example
sentiment, score = predict_sentiment("""
I actually created an account just so I could review this movie in the hopes of saving just one person from sitting through it and wasting three hours of their life. From the nonsensical timeframe to the ham, fisted political overtones to the extremely stupid ending. This went from being just a bad movie to perhaps the worst one I've ever seen. The only high point was a Steely Dan song in the soundtrack. Just listen to dirty work on Amazon music and save yourself the time, money and annoyance.

                                                                                                              """)
print(f"Sentiment: {sentiment} (Confidence: {score:.2f})")
```
