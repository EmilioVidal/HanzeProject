Saved Model Test

```python
import spacy

# Load the saved model
nlp = spacy.load("movie_sentiment_model")
```

Analyze a anew review

```python
def predict_sentiment(text):
    doc = nlp(text)
    sentiment = max(doc.cats, key=doc.cats.get)
    score = doc.cats[sentiment]
    return sentiment, score

# Example
sentiment, score = predict_sentiment("""
This movie was not what I thought it was gonna be. This animated movie is an absolute disaster from start to finish. This animated movie is a forgettable mess. The pacing is all over the place, with scenes dragging endlessly and others rushing by without any proper setup. This uneven flow makes it difficult to stay engaged, as the movie struggles to find any rhythm or momentum. It lacks originality, creativity, and heart, making it one of the worst animated films in recent memory. Watching it feels like a waste of time, with nothing redeeming to hold onto
                                    """)
print(f"Sentiment: {sentiment} (Confidence: {score:.2f})")
```
